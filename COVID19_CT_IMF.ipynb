# Store CSV into DataFrame

```python
csv_file = "COVID clinical trials.csv"
clinical_data_df = pd.read_csv(csv_file, encoding='utf-8')
xls_file = "imf-dm-export-20200829.xls"
imf_data_df = pd.read_excel(xls_file)
clinical_data_df.head()
```

# Transform Clinical Trial DataFrame

```python
clinical_data_df.columns
```

```python
clinical_data_df.isna().sum()
```

```python
reduced_ct_df = clinical_data_df[["NCT Number", "Study Type", "Start Date", "Completion Date", "Locations"]]
```

```python
reduced_ct_df.head()
reduced_ct_df.shape
```

```python
reduced_ct_df1= reduced_ct_df.dropna(how="any")
```

```python
reduced_ct_df1.reset_index(inplace=True)
reduced_ct_df1.shape
```

```python
reduced_ct_df1.head()
```

```python
s1 = reduced_ct_df1["Locations"].str.split('|',n=1, expand=True)[0]
s1
```

```python
s2=s1.str.rsplit(',',n=1, expand =True)[1]
s2.unique()
```

```python
cleaned_df = pd.concat([reduced_ct_df1, s2], axis=1, join='inner')
cleaned_df.head()
```

```python
cleaned_df1= cleaned_df[["NCT Number", "Study Type", "Start Date", "Completion Date", 1]]
cleaned_df1.columns
```

```python
renamed_df1 = cleaned_df1.rename(columns= {1: "Country"})

renamed_df1.head()
```

```python
renamed_df1.info
```

# Transform GDP DataFrame

```python
# xls_file = "imf-dm-export-20200829.xls"
# imf_data_df = pd.read_excel(xls_file)
imf_data_df.head()
```

```python
imf_data_df.columns
```

```python
reduced_imf = imf_data_df[['Real GDP growth (Annual percent change)', 2016, 2017, 2018, 2019, 2020]]
```

```python
reduced_imf.head()
```

```python
rename_imf = reduced_imf.rename(columns={"Real GDP growth (Annual percent change)": "Country"})
rename_imf.head()
```

```python
rename_imf.isna().sum()
```

```python
cleaned_imf_df = rename_imf.dropna(how="any")
```

```python
cleaned_imf_df.head()
```

```python
cleaned_imf_df["Country"].unique()
```

```python
# changed = {'China' : 'China', ' Hong Kong' : 'Hong Kong SAR'}
# for (row, col) in renamed_df1.iterrows():	
#     if (col.Country) in changed:	
#          renamed_df1['Country'].replace(to_replace=col.Country, value=dict.value, inplace =True)	
```

```python
changed = {'China' : "China, People's Republic of", 'Hong Kong' : 'Hong Kong SAR', 'Islamic Republic of' : 'Iran', 'North Macedonia' : 'North Macedonia ', 'Republic of' : 'Korea, Republic of', 'Taiwan' : 'Taiwan Province of China','The Democratic Republic of the' : 'Congo, Dem. Rep. of the'}
renamed_df1["Country"]=renamed_df1["Country"].str.strip()
for value, rep_value in changed.items():
    renamed_df1['Country'].replace(to_replace=value, value=rep_value, inplace =True)
    
```

```python
renamed_df1["Country"].unique()
```

# Merge Cleaned DataFrame

```python
covid19_CT_GDP= pd.merge(renamed_df1, cleaned_imf_df, on= ["Country", "Country"])
```

```python
covid19_CT_GDP.head()
```

# Create database connection

```python
rds_connection_string = "project:project@localhost:5432/covid19_CT_GDP_db"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

## Check for Table

```python
engine.table_names()
```

## Use pandas to load csv converted DataFrame into database

```python
covid19_CT_GDP.to_sql(name='covid_CT_GDP', con=engine, if_exists='append', index=False)
```
